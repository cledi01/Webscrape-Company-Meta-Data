```python
unique_domains = ["abc.net.au", "abqjournal.com", "adelaidenow.com.au", "espncricinfo.com", "www.audi.de"]
```

```python
# Navigate to search list and then navigate to specific company page
domain_list = []
url_list = []
company_name_list = []
number_of_employees_list = []
parent_company_list = [] 
annual_sales_list = []
for url_entry in unique_domains:
    try:
        driver.get(f"https://app-avention-com.ezproxy.library.tufts.edu/search/results/company?q={url_entry}")
        try:
            search_item = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "results-name"))
            )
        except:
            continue
        # search_list = driver.find_element_by_class_name("search-results-container")
        # search_item = search_list.find_element_by_class_name("results-name")
        company_site = search_item.get_attribute("href")
        driver.get(company_site)
        # time.sleep(3)
        try:
            url_span = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CLASS_NAME, "rpt-company-primaryurl"))
                )
        except:
            continue
        url_text = url_span.find_element_by_tag_name("a").text
        # Validate URL
        if (re.search(url_entry, url_text) == None):
            continue
        # Note this doesn't differentiate between actual and total

        name_div = driver.find_element_by_class_name("rpt-company-name")
        company_name = name_div.find_element_by_tag_name("span").text

        number_of_employees = -1
        try:
            tags = driver.find_elements_by_class_name("rpt-company-employees-this-site")
            texts = [tag.text.split(" ") for tag in tags]
            for text in texts:
                if text[1] == "(Total)":
                    number_of_employees = int(str.replace(text[0], ',' ,''))
                    break
        except Exception as e:
            print(e)
            pass

        parent_company = "NA"
        try:
            parent_div = driver.find_element_by_class_name("rpt-company-parent-company-name")
            parent_company = parent_div.find_element_by_tag_name("a").text
        except:
            pass

        annual_sales = driver.find_element_by_class_name("rpt-company-annual-sales").text
        # TODO maybe do company types?
    #     company_type = "NA"
    #     try:
        domain_list.append(url_entry)
        url_list.append(url_text)
        company_name_list.append(company_name)
        number_of_employees_list.append(number_of_employees)
        parent_company_list.append(parent_company)
        annual_sales_list.append(annual_sales)
    except Exception as e: 
        print(e)
        
data = {
    "domain": domain_list,
    "scraped_url": url_list,
    "company_name": company_name_list,
    "num_employees": number_of_employees_list,
    "parent_company": parent_company_list,
    "annual_sales": annual_sales_list
}
df_result = pd.DataFrame(data)
df_result.to_csv("result.csv", index=False)
```
